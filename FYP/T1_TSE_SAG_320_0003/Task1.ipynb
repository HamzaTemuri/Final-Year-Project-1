```python
def read_(i):
    ds = pydicom.dcmread(i)
    plt.imshow(ds.pixel_array,cmap=plt.cm.bone)
```

```python
read_(list[7])
```

```python
def read_crop(j):
    ds = pydicom.dcmread(j)
    plt.imshow(ds.pixel_array,cmap=plt.cm.bone)
    cropped = ds.pixel_array[:, 90:270]
    plt.imshow(cropped,cmap=plt.cm.bone);
```

```python
read_crop(list[7])
```
