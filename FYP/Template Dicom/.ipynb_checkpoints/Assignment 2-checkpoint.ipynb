```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

im = np.array(Image.open('4.jpg'), dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((2361,1471),669,567,linewidth=2,edgecolor='r',facecolor='none',angle=180)

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
