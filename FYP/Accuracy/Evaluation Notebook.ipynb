```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('testc.csv')
```

### Confusion Matrix 

```python
df.loc[(df['Class'] == 1) & (df['Class_R'] == 1) , 'TP'] = 1 
df.loc[(df['Class'] == 0) & (df['Class_R'] == 1) , 'FP'] = 1
df.loc[(df['Class'] == 1) & (df['Class_R'] == 0) , 'FN'] = 1
df.loc[(df['Class'] == 0) & (df['Class_R'] == 0) , 'TN'] = 1
```

```python
tp = int(df['TP'].sum())
fp = int(df['FP'].sum())
fn = int(df['FN'].sum())
tn = int(df['TN'].sum())
print('TRUE POSITIVE :',tp)
print('TRUE NEGATIVE :',tn)
print('FALSE POSITIVE:',fp)
print('FALSE NEGATIVE:',fn)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = ['TN', 'TP', 'FP', 'FN']
y = [tn,tp,fp,fn]
ax.bar(x,y,width=0.5,color = 'gbyr')
plt.show()
```

```python
# Creating dataset 
cm = ['True Negative', 'True Positive', 'False Positive', 'False Negative'] 
  
data = [tn,tp,fp,fn]
  
# Creating plot 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cm) 
  
# show plot 
plt.show() 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(df['Class'], df['Class_R'])

cm_display = ConfusionMatrixDisplay(cm).plot()
```

### Accuracy:

```python
#ACCURACY:
Accuracy = (tp + tn)/(tp + tn + fp + fn)
print("ACCURACY: {0}%".format(int(Accuracy*100)))
```

### Precision:

```python
#PRECISION:
Precision = tp/(tp + fp)
print("PRECISION: {0}%".format(int(Precision*100)))
```

### Recall:

```python
#RECALL:
Recall = tp/(tp + fn)
print("RECALL: {0}%".format(int(Recall*100)))
```

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = ['Accuracy', 'Precision', 'Recall']
y = [Accuracy*100,Precision*100,Recall*100]
ax.bar(x,y,color=['violet', 'orange', 'cyan'], width=0.5, edgecolor='white')
plt.show()
```
