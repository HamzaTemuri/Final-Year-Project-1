```python
def read_(i):
    ds = mpimg.imread(i)
    plt.imshow(ds,cmap=plt.cm.bone)
```

```python
def read_crop(j):
    ds = mpimg.imread(j)
    plt.imshow(ds,cmap=plt.cm.bone)
    cropped = ds[50:345, 125:340]
    plt.imshow(cropped,cmap=plt.cm.bone);
```

```python
read_crop(lists[2])
```

```python
def crop_(j):
    return j[50:345, 125:340]
```

```python
# Save Images Fn
def save_(img,name,path):
    loc = os.path.join(path, name)
    norm = (img - np.min(img)) / (np.max(img) - np.min(img))
    matplotlib.image.imsave(loc,crop_(norm),cmap=plt.cm.bone)
```
